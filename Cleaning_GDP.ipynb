```python
# Import dependencies
import pandas as pd
import numpy as np
import plotly.express as plt
```

```python
# import dataset
gdp_df = pd.read_csv('Resources/US_GDPbyState.csv')
gdp_df
```

```python
# Group by All Industry Total
gdp_df2 = gdp_df.loc[gdp_df['Description'] == "All industry total"]
gdp_df2
```

```python
# Drop unneccessary columns
gdp_df2 = gdp_df2.drop(['IndustryClassification', 'Description', 'Unit'], axis = 1)
gdp_df2
```

```python
# Stack Years into a column
gdp_df3 = gdp_df2.set_index(['GeoName']).stack()\
    .reset_index(name = "GDP").rename(columns={'level_1':'Year', 'GeoName':'State'})
gdp_df3
```

```python
# Replace "United States *" with "United States"
gdp_df3['State'] = gdp_df3['State'].str.replace(r'United States \*', r'United States')

gdp_df3
```

```python
# Multiply GDP by 1000 (in excel) to get actual GDP numbers
gdp_df3.to_csv('Resources/US_GDPbyState1.csv')
```

```python
# Re-import data and import state abbreviations
gdp_df4 = pd.read_csv('Resources/US_GDPbyState1.csv')
state_abbrev = pd.read_csv('Resources/StateAbbrev.csv')
print(gdp_df4)
print(state_abbrev)
```

```python
#Add abbrev column to gdp_df
gdp_df4['state_lower'] = gdp_df4['State'].str.lower()
state_abbrev['state_lower'] = state_abbrev['State'].str.lower()
gdp_df_cleaned = gdp_df4.merge(state_abbrev, on="state_lower", how="inner")\
    .drop(['State_y', 'state_lower'], axis = 1)\
    .rename(columns ={'State_x':'State'})
gdp_df_cleaned = gdp_df_cleaned[['State', 'Abbrev', 'Year', 'GDP']]
gdp_df_cleaned
```

```python
# Export cleaned data to csv
gdp_df_cleaned.to_csv('Resources/US_GDPbyState_Cleaned.csv')
```
