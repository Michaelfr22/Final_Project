```python
# Import pandas
import pandas as pd
```

```python
# Import data
pop_df = pd.read_csv('Resources/US_PopbyState.csv')
pop_df
```

```python
# Drop 'Totals' rows
pop_df = pop_df.dropna()
```

```python
# Rename Year column
pop_df['Year'] = pop_df['Yearly July 1st Estimates']
pop_df.drop('Yearly July 1st Estimates', axis = 1, inplace = True)
pop_df = pop_df[['State', 'Year', 'Population']]
pop_df
```

```python
# Removing decimal from year column
pop_df.style.set_precision(0)
```

```python
# Export cleaned csv
pop_df.to_csv('Resources/US_PopbyState_Cleaned.csv')
```
